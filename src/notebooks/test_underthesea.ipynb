Part-of-Speech (POS) Tags:

R (Phụ từ / Adverb or Particle): "Chưa"
V (Động từ / Verb): "tiết lộ", "lịch trình"
E (Giới từ / Preposition): "tới", "của"
Np (Danh từ riêng / Proper Noun): "Việt Nam", "Meox", "Do Nam", "Chung"
N (Danh từ / Noun): "Tổng thống"

Chunking Tags (BIO Format):

B-VP: Begin of a Verb Phrase
B-PP: Begin of a Prepositional Phrase
B-NP: Begin of a Noun Phrase
O: Outside of any chunked phrase

Named Entity Recognition (NER) Tags (BIO Format):

B-LOC: Begin of a Location (e.g., "Việt Nam", "Meox")
B-PER: Begin of a Person's name (e.g., "Do Nam")
I-PER: Inside / Continuation of a Person's name (e.g., "Chung")
O: Outside of any Named Entity (not a named entity)

```python
from underthesea import word_tokenize, ner, sentiment

word_tokenize("Chàng trai 9X Quảng Trị khởi nghiệp từ nấm sò")
# ["Chàng trai", "9X", "Quảng Trị", "khởi nghiệp", "từ", "nấm", "sò"]

outner = ner("Chưa tiết lộ lịch trình tới Việt Nam của Tổng thống Mỹ Donald Trump")
# [... ('Việt Nam', 'Np', 'B-NP', 'B-LOC'), ... ('Donald', 'Np', 'B-NP', 'B-PER'), ('Trump', 'Np', 'B-NP', 'I-PER')]

sentiment("Sản phẩm hơi nhỏ nhưng chất lượng tốt, đóng gói cẩn thận.")
# 'positive'
```
